```python
class Producto:
    def __init__(self, referencia, nombre, pvp, descripcion):
        self.referencia = referencia
        self.nombre = nombre
        self.pvp = pvp
        self.descripcion = descripcion
    
    def __str__(self):
        return """\
REFERENCIA      {}
NOMBRE          {}
PVP             {}
DESCRIPCIÓN     {}""".format(self.referencia, self.nombre, self.pvp, self.descripcion)

class Adorno(Producto):
    pass

class Alimento(Producto):
    productor = ""
    distribuidor = ""

    def __str__(self):
        return """\
REFERENCIA      {}
NOMBRE          {}
PVP             {}
DESCRIPCIÓN     {}
PRODUCTOR       {}
DISTRIBUIDOR    {}""".format(self.referencia, self.nombre, self.pvp, self.descripcion, self.productor, self.distribuidor)

class Libro(Producto):
    isbn = ""
    autor = ""

    def __str__(self):
        return """\
REFERENCIA      {}
NOMBRE          {}
PVP             {}
DESCRIPCIÓN     {}
ISBN            {}
AUTOR           {}""".format(self.referencia, self.nombre, self.pvp, self.descripcion, self.isbn, self.autor)

p = Producto("00000A", "Bolígrafo", 16.45, "Un bolígrafo muy sencillo")
a = Adorno("00000B", "Vaso Adornado", 3.98, "Un Vaso adornado con dibujos de vacas")

al = Alimento("00000t", "Chocolate Nestle", 4.50, "Chocolate chocolateado")
al.productor = "Nestle"
al.distribuidor = "Distribuidor Nestle SA"

l = Libro("00009V", "La cenicienta", 10.30, "El libro famoso de la cenicienta")
l.isbn = "0000-45423-HBKJ0"
l.autor = "Hector Acosta"

print(p)
print(a)
print(al)
print(l)
```
